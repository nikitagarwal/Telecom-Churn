## Filtering the high value customer 

### Good Phase 
- 80% of the revenue is generated by the top 20% of customers 
- So we will filter the data for high value customer from the 6th and the 7Th month as that is the good phase 

```python
df_churn.head(10)
```

### 1) Calculate the total Data recharge amount

```python
#calculate the data recharge amount for the month 6th and 7th 
df_churn["total_data_recharge_amt_6"]=df_churn["av_rech_amt_data_6"]*df_churn["total_rech_data_6"]
df_churn["total_data_recharge_amt_7"]=df_churn["av_rech_amt_data_7"]*df_churn["total_rech_data_7"]
```

### 2) Calculate total recharge amount 
(call recharge amount + Data recharge amount )

```python
#calculate the toatl rechareg amount for the month 6th and 7th
df_churn["total_recharge_amt6"]=df_churn["total_rech_amt_6"]+df_churn["total_data_recharge_amt_6"]
df_churn["total_recharge_amt7"]=df_churn["total_rech_amt_7"]+df_churn["total_data_recharge_amt_7"]
```

### 3) To Calculate the average recharge done by the customer in June and July

```python
# Calculate the average recharge amount 
df_churn["avg_rech_amt"]=(df_churn["total_recharge_amt6"]+df_churn["total_recharge_amt7"])/2
```

### 4) 70th percentile recharge amount 

```python
df_cutoff=df_churn["avg_rech_amt"].quantile(0.7)
df_cutoff
```

- we have the value that is 478 as the 70th percentile 
- The customers above this value will be considered as High Value customers

### 5)Filtered the data based on the 70th percentile of the recharge amount 

```python
#filtering the data based on the cutoff value obtained from the 70th percentile 
df_churn_filtered =df_churn[df_churn["avg_rech_amt"] >=  df_cutoff]
```

```python
#checking the shape of the data filtered on the 70th percentile basis 
df_churn_filtered.shape
```

- we have 30001 filtered rows and 212 columns left of the high value customers

## Derive Churn

## Rescaling the Features
As we have seen in the demonstration for Simple Linear Regression, scaling doesn't impact your model. So it is extremely important to rescale the variables so that they have a comparable scale. If we don't have comparable scales, then some of the coefficients as obtained by fitting the regression model might be very large or very small as compared to the other coefficients. This might become very annoying at the time of model evaluation. So it is advised to use standardization or normalization so that the units of the coefficients obtained are all on the same scale. As you know, there are two common ways of rescaling:

- Min-Max scaling
- Standardisation (mean-0, sigma-1)

This time, we will use Standardisation scaling.
